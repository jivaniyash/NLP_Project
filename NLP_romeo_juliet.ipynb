Create a TextBlob

```python
from textblob import TextBlob
```

```python
text = 'Yesterday was a beautiful day. Tomorrow looks like bad weather.'
blob_txt = TextBlob(text)
blob_txt
```

```python
#tokenize text into sentences
blob_txt.sentences

#[Sentence("Yesterday was a beautiful day."),
# Sentence("Tomorrow looks like bad weather.")]
```

```python
#tokenize text into words
blob_txt.words

# WordList(['Yesterday', 'was', 'a', 'beautiful', 'day', 'Tomorrow', 'looks', 'like', 'bad', 'weather'])
```

```python
# list of (word, part-of-speech-tag) tuples
blob_txt.tags

# [('Yesterday', 'NN'),
#  ('was', 'VBD'),
#  ('a', 'DT'),
#  ('beautiful', 'JJ'),
#  ('day', 'NN'),
#  ('Tomorrow', 'NNP'),
#  ('looks', 'VBZ'),
#  ('like', 'IN'),
#  ('bad', 'JJ'),
#  ('weather', 'NN')]
```

```python
# Extracting Noun Phrases
# WordList of Word objects
blob_txt.noun_phrases 

# WordList(['yesterday', 'beautiful day', 'tomorrow', 'bad weather'])
```

```python
# polarity & sentiment of the text
blob_txt.sentiment

# Sentiment(polarity=0.07500000000000007, subjectivity=0.8333333333333333)
```

```python
# polarity & sentiment of each of the sentence in the text
for sentence in blob_txt.sentences:
    print(f'{sentence}\n{sentence.sentiment}\n')

# Yesterday was a beautiful day.
# Sentiment(polarity=0.85, subjectivity=1.0)

# Tomorrow looks like bad weather.
# Sentiment(polarity=-0.6999999999999998, subjectivity=0.6666666666666666)
```

Word Frequencies via 'word_counts' Dictionary in a TextBlob

```python
from pathlib import Path
```

[RomeoAndJuliet.txt](https://www.gutenberg.org/ebooks/1513) is a book available in [gutenberg.org](https://www.gutenberg.org/).

This book is uesd for analysis of frequency and other uses.

```python
# text preprocessing
# removing content of headers & footers
rj_book_edited = open('rj_book_edited.txt','w')
with open('RomeoAndJuliet.txt','r') as rj:
    for line in rj:
        if line.startswith('*** START OF'):
            break
    # creating new file : rj_book_edited after removing header & footer text of the book
    for line in rj:
        if line.startswith('*** END OF'):
            break
        line = line.lower()
        # line = re.sub("[^A-z \n.]", " ",line)
        line = line.replace('\n',' ') 
        rj_book_edited.write(line)
```

```python
# load the text file in TextBlob object
rj_book = TextBlob(Path('rj_book_edited.txt').read_text())
rj_book.sentiment
```

```python
# comparing count of words - romeo & juliet in book
rj_book.word_counts['romeo']
rj_book.word_counts['juliet'] 
```

Deleting Stopwords - less significant words

```python
from nltk.corpus import stopwords
from operator import itemgetter
```

```python
waste_words = stopwords.words('english') # wastewords
```

```python
sample = TextBlob('Today is not a great day to go for a ride')

# removing waste_words from sample 
new_list = []
for word in sample.word_counts:
    if word in waste_words:
        continue
    new_list.append(word.lower())
new_list # ['today', 'great', 'day', 'go', 'ride']
```

```python
new_list = []
for word,count in rj_book.word_counts.items():
    if word in waste_words:
        continue
    new_list.append((word,count))

f_list = sorted(new_list, key=itemgetter(1), reverse=True)

# top 20 frequency words
top20 = f_list[:20]
top20
```

Crearting Pandas Dataframe object for visualization

```python
import pandas as pd
```

```python
data_frame = pd.DataFrame(top20,columns=['word','count'])
data_frame
```

```python
import matplotlib.pyplot as plt
```

```python
axes = data_frame.plot.bar(x='word', y='count')
plt.gcf().tight_layout() 
```
